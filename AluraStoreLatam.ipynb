# 2. Ventas por categoría

```python
# prompt: calcular la cantidad de productos vendidos por categoria en cada tienda.

# Agrupar por categoría y contar las ventas para cada tienda
ventas_por_categoria_tienda1 = tienda.groupby("Categoría del Producto")["Precio"].count()
ventas_por_categoria_tienda2 = tienda2.groupby("Categoría del Producto")["Precio"].count()
ventas_por_categoria_tienda3 = tienda3.groupby("Categoría del Producto")["Precio"].count()
ventas_por_categoria_tienda4 = tienda4.groupby("Categoría del Producto")["Precio"].count()

# Imprimir los resultados
print("Tienda 1:")
print(ventas_por_categoria_tienda1)
print("\nTienda 2:")
print(ventas_por_categoria_tienda2)
print("\nTienda 3:")
print(ventas_por_categoria_tienda3)
print("\nTienda 4:")
ventas_por_categoria_tienda4
```

```python
# prompt: La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.

# Encontrar las categorías más populares en cada tienda

def top_categories(df, n=5):
    """
    Encuentra las n categorías más populares en un DataFrame dado.

    Args:
        df: DataFrame de pandas con una columna 'Categoría del Producto'.
        n: Número de categorías principales para devolver (predeterminado es 2).

    Returns:
        Una serie de pandas con las n categorías principales y sus recuentos.
    """

    # Agrupar por categoría y contar las ventas
    category_counts = df.groupby("Categoría del Producto")["Precio"].count()

    # Ordenar por conteo descendente y tomar las n categorías superiores
    top_n_categories = category_counts.sort_values(ascending=False).head(n)

    return top_n_categories


# Aplicar la función a cada tienda
top_categories_tienda1 = top_categories(tienda)
top_categories_tienda2 = top_categories(tienda2)
top_categories_tienda3 = top_categories(tienda3)
top_categories_tienda4 = top_categories(tienda4)


# Imprimir los resultados
print("Categorías más populares (Top 5) de cada tienda:")
print("\nTienda 1:")
print(top_categories_tienda1)
print("\nTienda 2:")
print(top_categories_tienda2)
print("\nTienda 3:")
print(top_categories_tienda3)
print("\nTienda 4:")
top_categories_tienda4
```

# 3. Calificación promedio de la tienda


En este paso, debes calcular las calificaciones promedio de los clientes para cada tienda. El objetivo es conocer la satisfacción del cliente con los productos vendidos.

```python
# prompt: calcular las calificaciones promedio de los clientes para cada tienda. El objetivo es conocer la satisfacción del cliente con los productos vendidos.

# Calcular la calificación promedio para cada tienda
promedio_tienda1 = tienda["Calificación"].mean()
promedio_tienda2 = tienda2["Calificación"].mean()
promedio_tienda3 = tienda3["Calificación"].mean()
promedio_tienda4 = tienda4["Calificación"].mean()

# Imprimir los resultados
print(f"Calificación promedio de la tienda 1: {promedio_tienda1}")
print(f"Calificación promedio de la tienda 2: {promedio_tienda2}")
print(f"Calificación promedio de la tienda 3: {promedio_tienda3}")
print(f"Calificación promedio de la tienda 4: {promedio_tienda4}")

# Crear un DataFrame para visualizar mejor los resultados
import pandas as pd
data = {'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
        'Calificación Promedio': [promedio_tienda1, promedio_tienda2, promedio_tienda3, promedio_tienda4]}
df_promedios = pd.DataFrame(data)

# Mostrar el DataFrame
print("\nDataFrame con calificaciones promedio:")
print(df_promedios)

# Gráfico de barras para comparar calificaciones promedio
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.bar(df_promedios['Tienda'], df_promedios['Calificación Promedio'], color=['skyblue', 'lightcoral', 'lightgreen', 'lightsalmon'])
plt.xlabel("Tienda")
plt.ylabel("Calificación Promedio")
plt.title("Calificaciones Promedio por Tienda")
plt.ylim(0, 5) # Ajustar el límite del eje y para una mejor visualización
plt.show()
```

# 4. Productos más y menos vendidos

En este paso, debes identificar qué productos fueron los más vendidos y los menos vendidos en cada tienda. Visualiza los resultados para que quede claro qué productos destacaron en ventas en cada tienda.

```python
# prompt: identificar qué productos fueron los más vendidos y los menos vendidos en cada tienda.

# Función para encontrar los productos más y menos vendidos
def top_bottom_products(df, n=5):
    """
    Encuentra los n productos más y menos vendidos en un DataFrame dado.

    Args:
        df: DataFrame de pandas con columnas 'Producto' y 'Precio'.
        n: Número de productos principales y menos vendidos para devolver (predeterminado es 5).

    Returns:
        Una tupla que contiene dos series de pandas:
        - top_products: Los n productos más vendidos y sus recuentos.
        - bottom_products: Los n productos menos vendidos y sus recuentos.
    """

    # Agrupar por producto y contar las ventas
    product_counts = df.groupby("Producto")["Precio"].count()

    # Ordenar por conteo descendente y tomar los n productos superiores e inferiores
    top_products = product_counts.sort_values(ascending=False).head(n)
    bottom_products = product_counts.sort_values(ascending=True).head(n)

    return top_products, bottom_products


# Aplicar la función a cada tienda
top_tienda1, bottom_tienda1 = top_bottom_products(tienda)
top_tienda2, bottom_tienda2 = top_bottom_products(tienda2)
top_tienda3, bottom_tienda3 = top_bottom_products(tienda3)
top_tienda4, bottom_tienda4 = top_bottom_products(tienda4)

# Imprimir los resultados
print("Productos más y menos vendidos (Top 5) de cada tienda:")

print("\nTienda 1:")
print("Más vendidos:")
print(top_tienda1)
print("\nMenos vendidos:")
print(bottom_tienda1)

print("\nTienda 2:")
print("Más vendidos:")
print(top_tienda2)
print("\nMenos vendidos:")
print(bottom_tienda2)

print("\nTienda 3:")
print("Más vendidos:")
print(top_tienda3)
print("\nMenos vendidos:")
print(bottom_tienda3)

print("\nTienda 4:")
print("Más vendidos:")
print(top_tienda4)
print("\nMenos vendidos:")
bottom_tienda4
```

```python
# prompt: Visualiza los resultados para que quede claro qué productos destacaron en ventas en cada tienda.

import matplotlib.pyplot as plt

# ... (Your existing code) ...

# Function to plot top and bottom products
def plot_top_bottom_products(top_products, bottom_products, title):
    fig, axes = plt.subplots(1, 2, figsize=(16, 6))  # Create subplots for better visualization

    # Plot top products
    axes[0].bar(top_products.index, top_products.values)
    axes[0].set_title(f'{title} - Productos Más Vendidos')
    axes[0].set_xlabel('Producto')
    axes[0].set_ylabel('Cantidad Vendida')
    axes[0].tick_params(axis='x', rotation=45, labelsize=8) # Rotate x-axis labels for readability


    # Plot bottom products
    axes[1].bar(bottom_products.index, bottom_products.values)
    axes[1].set_title(f'{title} - Productos Menos Vendidos')
    axes[1].set_xlabel('Producto')
    axes[1].set_ylabel('Cantidad Vendida')
    axes[1].tick_params(axis='x', rotation=45, labelsize=8) # Rotate x-axis labels


    plt.tight_layout()  # Adjust layout to prevent overlapping labels
    plt.show()

# Plot for each store
plot_top_bottom_products(top_tienda1, bottom_tienda1, "Tienda 1")
plot_top_bottom_products(top_tienda2, bottom_tienda2, "Tienda 2")
plot_top_bottom_products(top_tienda3, bottom_tienda3, "Tienda 3")
plot_top_bottom_products(top_tienda4, bottom_tienda4, "Tienda 4")
```

# 5. Envío promedio por tienda

En este paso, debes calcular el costo de envío promedio para cada tienda. El objetivo es comprender cuánto se gasta, en promedio, en el envío de cada tienda.

```python
# prompt: calcular el costo de envío promedio para cada tienda.

# Calcular el costo de envío promedio para cada tienda
promedio_envio_tienda1 = tienda["Costo de envío"].mean()
promedio_envio_tienda2 = tienda2["Costo de envío"].mean()
promedio_envio_tienda3 = tienda3["Costo de envío"].mean()
promedio_envio_tienda4 = tienda4["Costo de envío"].mean()

# Imprimir los resultados
print(f"Costo de envío promedio de la tienda 1: {promedio_envio_tienda1}")
print(f"Costo de envío promedio de la tienda 2: {promedio_envio_tienda2}")
print(f"Costo de envío promedio de la tienda 3: {promedio_envio_tienda3}")
print(f"Costo de envío promedio de la tienda 4: {promedio_envio_tienda4}")

# Crear un DataFrame para visualizar mejor los resultados
data_envio = {'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
        'Costo de Envío Promedio': [promedio_envio_tienda1, promedio_envio_tienda2, promedio_envio_tienda3, promedio_envio_tienda4]}
df_promedios_envio = pd.DataFrame(data_envio)

# Mostrar el DataFrame
print("\nDataFrame con costos de envío promedio:")
print(df_promedios_envio)

# Gráfico de barras para comparar costos de envío promedio
plt.figure(figsize=(8, 6))
plt.bar(df_promedios_envio['Tienda'], df_promedios_envio['Costo de Envío Promedio'], color=['skyblue', 'lightcoral', 'lightgreen', 'lightsalmon'])
plt.xlabel("Tienda")
plt.ylabel("Costo de Envío Promedio")
plt.title("Costos de Envío Promedio por Tienda")
plt.show()
```
